```python
# Answer:


# Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in Python. It provides a wide variety of plotting functions and tools for creating high-quality plots, charts, and graphs for data analysis and visualization tasks.

# Matplotlib is commonly used for:

# Exploratory Data Analysis (EDA): Matplotlib allows users to quickly visualize datasets to understand their distributions, patterns, and relationships.

# Presentation Graphics: It is widely used for generating publication-quality figures and plots for scientific research papers, presentations, and reports.

# Custom Visualizations: Matplotlib provides extensive customization options, allowing users to create highly customizable plots tailored to their specific requirements.

# Integration with Scientific Computing Libraries: It integrates seamlessly with other Python libraries such as NumPy, Pandas, and SciPy, making it a powerful tool for visualizing data structures and results from scientific computations.

# Web-based Visualization: Matplotlib can be used in web applications through frameworks like Flask and Django, enabling the creation of interactive data visualizations for web-based data analysis platforms.


#Some of the commonly used plots that can be created using the Pyplot module of Matplotlib include:


# Line Plot: Used to visualize the relationship between two continuous variables by connecting data points with straight lines.

# Scatter Plot: Useful for visualizing the distribution and relationship between two continuous variables by displaying individual data points on a two-dimensional plane.

# Histogram: Represents the distribution of a single continuous variable by dividing the data into bins and displaying the frequency of observations in each bin as bars.

# Bar Plot: Displays categorical data with rectangular bars, where the height or length of each bar corresponds to the frequency or value of the category.

# Pie Chart: Represents categorical data as slices of a circular pie, where each slice's size represents the proportion of each category relative to the whole.
```

```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# import numpy as np
# np.random.seed(3)
# x = 3 + np.random.normal(0, 2, 50)
# y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
#Answer:

# A scatter plot is a type of plot used to visualize the relationship between two continuous variables.
# Each data point is represented as a dot on the plot, with the x-axis typically representing one variable and the y-axis representing the other variable.
# Scatter plots are useful for identifying patterns, trends, and correlations between variables.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plot scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Answer:
    
# The subplot() function in Matplotlib is used to create multiple plots within the same figure.
# It allows you to divide the figure into a grid of rows and columns and specify which subplot to plot your data in.
# This is particularly useful when you want to compare multiple datasets or visualize different aspects of your data side by side.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
# profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Answer:
    
# A bar plot is a type of plot that represents categorical data with rectangular bars. 
# The length or height of each bar corresponds to the value of the data it represents. 
# Bar plots are typically used to compare the values of different categories or groups.

# Bar plots are useful for:

# Comparing Categorical Data: Bar plots allow for a quick visual comparison of the values of different categories or groups.
# Showing Distribution: They can be used to show the distribution of a categorical variable.
# Visualizing Trends: Bar plots can reveal trends or patterns in categorical data.
# Highlighting Differences: They can highlight differences or similarities between categories.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.title('Profit of Companies')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for y-axis
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='salmon')
plt.title('Profit of Companies')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Add gridlines for x-axis
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)
```

```python
# Answer:
    

# A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. It is used to visualize the spread and skewness of the data along with identifying outliers.

# Box plots are useful for:

# Identifying Central Tendency: The central line in the box represents the median, providing insight into the central tendency of the data.
# Understanding Spread: The length of the box indicates the spread of the middle 50% of the data (interquartile range, IQR), while the whiskers show the full range of the data, excluding outliers.
# Detecting Outliers: Outliers, if present, are shown as individual points beyond the whiskers, making them easily identifiable.
# Comparing Distributions: Box plots allow for visual comparison of distributions between multiple datasets.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Data')
plt.xlabel('Box')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for y-axis
plt.show()
```
