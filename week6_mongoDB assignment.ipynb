```python
# Question 1: What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?

# Answer :

# MongoDB is a popular open-source NoSQL database system that stores data in a document-oriented format. Instead of using tables and rows like in traditional relational databases, MongoDB stores data in documents that are organized into collections. These documents can contain a variety of data types, such as arrays and nested documents, and can be easily updated and queried.These documents are stored in bson (binary JSON) format.

# Non-relational databases, also known as NoSQL databases, are databases that do not use the traditional table-based relational database model. Instead, they use different data models, such as key-value, document-oriented, or graph-based, to store and organize data. Non-relational databases are often used for handling large amounts of unstructured or semi-structured data, and they can be more flexible and scalable than traditional SQL databases.

# MongoDB is preferred over SQL databases in scenarios where:

# Flexibility and scalability are important: MongoDB is designed to be flexible and scalable, making it a good choice for handling large and complex data sets.

# Agile development: MongoDB's flexible schema allows for more agile and iterative development, as changes can be made to the database schema without requiring a complete redesign of the application.

# Big Data and real-time analytics: MongoDB is well-suited for big data and real-time analytics, as it can easily handle unstructured and semi-structured data.

# Cloud-based applications: MongoDB is often used for cloud-based applications, as it can easily scale to meet the demands of a growing user base.

# Geospatial data: MongoDB has built-in support for geospatial data, making it a popular choice for applications that involve location-based services.
```

```python
#Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
#demonstrate this.
```

```python

#In MongoDB, the find() method is used to query a collection and retrieve documents that match specified criteria. This method allows you to filter documents based on specific conditions and retrieve the results. Here's how you can use the find() method:

#Basic Query: You can pass an empty document {} to the find() method to retrieve all documents in the collection.

#Query with Conditions: You can specify conditions using a document with key-value pairs. MongoDB will return documents that match these conditions.

#Projection: You can specify which fields you want to include or exclude from the result set using projection. This helps in reducing the size of the returned documents.

#Sorting: You can sort the results based on one or more fields in ascending or descending order.

import pymongo


client = pymongo.MongoClient("mongodb+srv://ramc53112:ramc53112@cluster0.rxufn56.mongodb.net/?retryWrites=true&w=majority")


mydb = client["mydatabase"]
mycol = mydb["customers"]


data = [
    { "name": "John", "age": 30 },
    { "name": "Alice", "age": 25 },
    { "name": "Bob", "age": 35 },
    { "name": "Carol", "age": 40 }
]
mycol.insert_many(data)


print("All documents in the collection:")
for document in mycol.find():
    print(document)


print("\nDocuments with age greater than 30:")
for document in mycol.find({ "age": { "$gt": 30 } }):
    print(document)


print("\nName of all documents:")
for document in mycol.find({}, { "_id": 0, "name": 1 }):
    print(document)


print("\nDocuments sorted by age in descending order:")
for document in mycol.find().sort("age", pymongo.DESCENDING):
    print(document)
```

```python
# Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

#Answer:

# In MongoDB, the sort() method is used to specify the order in which the query results should be returned. This method allows you to sort the documents returned by a query based on one or more fields, either in ascending or descending order.

# The sort() method takes a document as an argument, where each field you want to sort by is a key, and the value of the key determines the sort order. If the value is 1, the field will be sorted in ascending order, and if the value is -1, the field will be sorted in descending order.
```

```python
import pymongo


client = pymongo.MongoClient("mongodb+srv://ramc53112:ramc53112@cluster0.rxufn56.mongodb.net/?retryWrites=true&w=majority")


mydb = client["mydatabase"]
mycol = mydb["customers"]


data = [
    { "name": "John", "age": 30 },
    { "name": "Alice", "age": 25 },
    { "name": "Bob", "age": 35 },
    { "name": "Carol", "age": 40 }
]
mycol.insert_many(data)


print("Documents sorted by age in ascending order:")
for document in mycol.find().sort("age", pymongo.ASCENDING):
    print(document)


print("\nDocuments sorted by age in descending order:")
for document in mycol.find().sort("age", pymongo.DESCENDING):
    print(document)
```

```python
#Q7. Explain why delete_one(), delete_many(), and drop() is used.
```

```python
#Answer
# In MongoDB, delete_one(), delete_many(), and drop() are used for different purposes related to removing data or collections from the database.

# delete_one():

# This method is used to delete a single document that matches the specified criteria.
# It takes a filter document as a parameter, which specifies the criteria for matching the document to be deleted.
# If multiple documents match the filter, only the first one encountered will be deleted.
# It is typically used when you want to remove a specific document from a collection.
# delete_many():

# This method is used to delete multiple documents that match the specified criteria.
# It takes a filter document as a parameter, similar to delete_one(), specifying the criteria for matching documents to be deleted.
# All documents that match the filter will be deleted.
# It is useful when you want to remove multiple documents from a collection based on certain criteria.
# drop():

# This method is used to drop an entire collection from the database.
# It does not take any parameters.
# Dropping a collection removes all documents within that collection and the collection itself.
# It is typically used when you want to completely remove a collection and all its data from the database.
# These methods provide flexibility in managing data within MongoDB collections:

# delete_one() and delete_many() allow for targeted removal of specific documents or groups of documents based on specified criteria.
# drop() provides a way to remove an entire collection and its contents in one operation.
# It's essential to use these methods carefully, especially drop(), as it permanently removes data from the database and cannot be undone. Always ensure you're deleting the correct data or collection to avoid unintended data loss.
```
