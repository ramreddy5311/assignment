```python
# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
# Answer:
    
# Seaborn is a powerful Python visualization library built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. 
# Here are five common plots that can be created using Seaborn, along with their typical uses:

# Scatter Plot: A scatter plot is used to visualize the relationship between two continuous variables. It helps in understanding the correlation or pattern between the variables.

# Histogram: Histograms are useful for visualizing the distribution of a single continuous variable. They show the frequency or count of observations falling within different bins.

# Bar Plot: Bar plots are used to compare categorical variables or to show the distribution of a categorical variable. They display the count, average, sum, or other statistical measures for each category.

# Box Plot: Box plots, also known as box-and-whisker plots, are used to visualize the distribution of a continuous variable and to identify any outliers. They provide a summary of the central tendency, spread, and skewness of the data.

# Violin Plot: Violin plots are similar to box plots but provide a more detailed representation of the data distribution. They display the density of the data at different values, offering insights into both central tendency and spread.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.
# Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
# Answer:
    

# To plot a line plot using the "fmri" dataset with Seaborn, you can follow these steps:

# Import the necessary libraries: Seaborn and Matplotlib.
# Load the "fmri" dataset using the load_dataset() function from Seaborn.
# Use the lineplot() function from Seaborn to create the line plot, specifying the "timepoint" column as the x-axis, the "signal" column as the y-axis, and using the "event" and "region" columns to differentiate between different lines.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Add title and labels
plt.title("FMRI Signal Over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.
# Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Age Distribution by Passenger Class')

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Fare Distribution by Passenger Class')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.
```

```python
# Answer:

# To plot a histogram for the 'price' column of the "diamonds" dataset from Seaborn and use the 'cut' column as the hue parameter, you can follow these steps:

# Import the necessary libraries: Seaborn and Matplotlib.
# Load the "diamonds" dataset using the load_dataset() function from Seaborn.
# Use the histplot() function from Seaborn to create the histogram, specifying the 'price' column as the data and the 'cut' column as the hue.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Reshape the dataset into a pivot table format
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data=flights_pivot, cmap='viridis', annot=True, fmt='d', linewidths=.5)
plt.title('Passenger Counts by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
