```python
# Q1. What is Flask Framework? What are the advantages of Flask Framework?

# Answer:

# Flask is a micro web framework written in Python. It is lightweight, easy to use, and designed to make it simple to build web applications quickly and with minimal boilerplate code. Flask is classified as a micro-framework because it does not require any particular libraries or tools, providing only the essentials needed for web development.

# Here are some key features and advantages of Flask:

# Lightweight and Minimalistic: Flask is minimalist in design, providing just the basics for building web applications. It does not impose any specific project structure or dependencies, giving developers the flexibility to choose the tools and libraries they prefer.

# Easy to Learn and Use: Flask's simplicity makes it easy for beginners to learn and use. Its intuitive API and clear documentation allow developers to quickly get started with building web applications without a steep learning curve.

# Extensible: Despite its minimalist design, Flask is highly extensible. It supports the use of extensions, which are third-party libraries that add additional functionalities to the framework, such as database integration, authentication, and form validation.

# Built-in Development Server: Flask comes with a built-in development server, making it easy to run and test web applications locally during development. This eliminates the need for setting up a separate web server during the early stages of development.

# Flexible URL Routing: Flask provides a simple and flexible mechanism for defining URL routes using decorators. This allows developers to map URL patterns to view functions easily, making it straightforward to create RESTful APIs and define application endpoints.

# Jinja2 Templating Engine: Flask integrates with the Jinja2 templating engine, which allows developers to create dynamic HTML templates using a syntax similar to Python. Jinja2 templates support inheritance, macros, and other powerful features for building reusable and maintainable UI components.

# RESTful Request Handling: Flask provides built-in support for handling HTTP requests in a RESTful manner, making it easy to implement CRUD (Create, Read, Update, Delete) operations for resources.

# Python Compatibility: As a Python web framework, Flask seamlessly integrates with Python libraries and tools, allowing developers to leverage the rich ecosystem of Python packages for various tasks such as database access, data manipulation, and third-party integrations.

# Overall, Flask's lightweight design, simplicity, extensibility, and compatibility with Python make it an excellent choice for building web applications, APIs, and prototypes quickly and efficiently.
```

```python
# Q3. What is App routing in Flask? Why do we use app routes?
```

```python
#Answer:
#In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to the corresponding Python functions, known as view functions, within a Flask application. 
# This is achieved using the @app.route() decorator.
# URL Handling: App routes allow us to define how different URLs should be handled within our Flask application. This enables us to create a structure for our web application and define what content or functionality should be provided at each URL.

# Organizing Code: App routes help in organizing code by associating specific URLs with corresponding view functions. This makes it easier to manage and maintain the codebase, especially as the application grows larger.

# Modularization: Flask allows us to modularize our application by defining routes in separate Python modules or files. This separation of concerns makes the codebase more maintainable and scalable.

# RESTful APIs: App routes are crucial for building RESTful APIs in Flask. Each route typically corresponds to a specific API endpoint, and the associated view function handles the HTTP request and response for that endpoint.
```

```python
# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
# url_for() function.
```

```python
#Answer:

# In Flask, the url_for() function is used for URL building. It generates a URL to a specific function based on its name. 
# This function accepts the name of the view function as an argument and returns the URL that corresponds to that function.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the index page'

@app.route('/about')
def about():
    return 'This is the about page'

@app.route('/contact')
def contact():
    return 'This is the contact page'

if __name__ == '__main__':
    with app.test_request_context():
        print("URL for index page:", url_for('index'))
        print("URL for about page:", url_for('about'))
        print("URL for contact page:", url_for('contact'))
```
